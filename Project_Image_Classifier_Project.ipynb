## Create Pipeline

````python
# TODO: Create a pipeline for each set.
# Create a pipeline for each set.
IMG_SIZE = 224
BATCH_SIZE = 64

def normalize_image(image, label):
    # Resize the image to the required size
    image = tf.cast(image, tf.float32)
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    # Normalize
    image = image / 255.0
    return image, label

#pipline complix calling
#training_batches = ds_train.shuffle(num_train_examples//4).map(normalize_image).batch(batch_size).prefetch(1)
#validation_batches = ds_val.map(normalize_image).batch(batch_size).prefetch(1)
#testing_batches = ds_test.map(normalize_image).batch(batch_size).prefetch(1)

#pipline function
def create_pipeline(dataset, shuffle=False, batch_size=BATCH_SIZE, buffer_size=None):
    if shuffle and buffer_size is not None:
        dataset = dataset.shuffle(buffer_size=buffer_size)
    dataset = dataset.map(normalize_image, num_parallel_calls=tf.data.AUTOTUNE)
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(buffer_size=tf.data.AUTOTUNE)
    return dataset

#test pipline
train_batches = create_pipeline(ds_train, shuffle=True, buffer_size=num_train_examples//4)
val_batches = create_pipeline(ds_val)
test_batches = create_pipeline(ds_test)
````

````python
# Build and train your network.
import tf_keras

tf.keras.backend.clear_session()
# TODO: Build and train your network.
URL = "https://tfhub.dev/google/imagenet/mobilenet_v2_140_224/feature_vector/4"

feature_extractor = hub.KerasLayer(URL, input_shape=(IMG_SIZE,IMG_SIZE,3), trainable=False)

model = tf_keras.Sequential()
model.add(feature_extractor)

# Add feed-forward layers
layer_neurons = [256, 128]
for neurons in layer_neurons:
    model.add(tf_keras.layers.Dense(neurons, activation='relu'))
    model.add(tf_keras.layers.Dropout(0.2))

# Add output layer
model.add(tf_keras.layers.Dense(num_classes, activation='softmax'))

model.summary()
````

````python
# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
EPOCHS = 30
# Train the model
history = model.fit(train_batches,  
                    epochs=EPOCHS, 
                    validation_data=val_batches)  
````

````python
# Plot the loss and accuracy values achieved during training for the training and validation set.
training_loss = history.history['loss']
validation_loss = history.history['val_loss']

training_accuracy = history.history['accuracy']
validation_accuracy = history.history['val_accuracy']

epochs_range = range(len(training_loss))

# Plot training & validation loss
plt.figure(figsize=(10, 8))

plt.subplot(121)
plt.plot(epochs_range, training_loss, label='Training Loss')
plt.plot(epochs_range, validation_loss, label='Validation Loss')
plt.title('training & validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Plot training & validation accuracy
plt.subplot(122)
plt.plot(epochs_range, training_accuracy, label='Training Accuracy')
plt.plot(epochs_range, validation_accuracy, label='Validation Accuracy')
plt.title('training & validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

#plt.tight_layout()
plt.show()
````

### Image Pre-processing

The `process_image` function should take in an image (in the form of a NumPy array) and return an image in the form of a NumPy array with shape `(224, 224, 3)`.

First, you should convert your image into a TensorFlow Tensor and then resize it to the appropriate size using `tf.image.resize`.

Second, the pixel values of the input images are typically encoded as integers in the range 0-255, but the model expects the pixel values to be floats in the range 0-1. Therefore, you'll also need to normalize the pixel values. 

Finally, convert your image back to a NumPy array using the `.numpy()` method.

````python
# Create the process_image function
def process_image(image):
    image = tf.convert_to_tensor(image, dtype=tf.float32)
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image /= 255.0
    image = image.numpy()
    return image
````
